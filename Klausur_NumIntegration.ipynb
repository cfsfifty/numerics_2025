```python
import sympy as sp
sp.init_printing()
from sympy import *
```

Numerische Integration, Aufgabe 1

```python
# (4;4),(6;7),(8;1),(10;6),(12;2)
x0=4; x1=6; x2=8; x3=10; x4=12;  
y0=4; y1=7; y2=1; y3=6;  y4=2;
h=Rational("8/4")
# Trapezregel
Trapez_1=h*(Rational("1/2")*Rational(y0)+Rational(y1)+Rational(y2)+Rational(y3)+Rational("1/2")*Rational(y4))
print(Trapez_1.evalf())
Trapez_1
```

```python
# Simpsonregel
Simp_1=Rational(h/3)*(Rational(y0)+Rational("4")*(Rational(y1)+Rational(y3))+Rational("2")*Rational(y2)+Rational(y4))
print(Simp_1.evalf())
Simp_1
```
