### Graphen Iterationsverfahren

```python
import math
import numpy as np              # array Funktionen
import matplotlib.pyplot as plt # plot Funktionen

pi= 3.14159265
x = np.arange(0, pi/2.0, 0.02) # x-Werte
y = np.cos(x)               # Zugehörige y-Werte der Wertetabelle
fig, axs = plt.subplots(1, 1) 
axs.plot(x, y, 'g-')       # Plot der Wertetabelle
axs.plot(x, x, 'g-')       # Plot der Wertetabelle
axs.plot(x, x-y, 'r-')       # Plot der Wertetabelle

axs.plot([0], [0], 'k.')
#axs.plot( [0, 0], [-2, 2], 'k-') 
axs.plot([0.739085, 0.739085], [-1.5,  1.5], 'k-') 
axs.plot([0, pi/2.0], [0,  0], 'k-') 
axs.plot([0, 0], [-1.5,  1.5], 'k-') 
axs.axis([0, pi/2.0, -1.5, 1.5])
axs.axis('equal')
fig.tight_layout()
```
