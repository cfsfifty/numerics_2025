Aufgabe 2: Interpolationspolynom 2.Grades

```python
a, b, c = symbols("a b c")
x   = Symbol("x")
px=a*x**2+b*x+c
glg1=px.subs({ x : -1 })-Rational(5)
glg2=px.subs({ x : 0})-Rational(2)
glg3=px.subs({ x : 1})-Rational(1)
glg3
```

```python
solve([glg1, glg2, glg3], dict=True)
```
