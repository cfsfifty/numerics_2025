```python
import sympy as sp
sp.init_printing()
from sympy import *

# Variable
x = Symbol("x")
# Ableitungen: f, df, df2, df3
f = 4*x**3-3*x**2+5*x-1
df  = diff(f, x)
print(df)
df2 = diff(df, x)
print(df2)
df3 = diff(df2, x)
print(df3)
df4 = diff(df3, x)
print(df4)
```

```python
f.subs(x, "2")
```

```python
df.subs(x, "2")
```

```python
df2.subs(x, "2")
```

```python
df3.subs(x, "2")
```

```python
# Eingabe-nahe Form
Tx=f.subs(x, "2")+df.subs(x, "2")/1*(x-2)+df2.subs(x, "2")/2*(x-2)**2+df3.subs(x, "2")/6*(x-2)**3
Tx
```

```python
# Ausmultipliziert
simplify(Tx)
```
