Aufgabe 1: Numerische Integrationsverfahren

```python
import math
import numpy as np              # array Funktionen
import matplotlib.pyplot as plt # plot Funktionen

h  = (2-0)/4
xx = np.arange(0, 2.5, 0.1) # x-Werte
yy = 3*math.e**(-xx) # y = f(x)
plt.plot(xx, yy, 'g-')                 # Plot 

plt.axis([0, 2.5, 0, 4])
plt.grid(True)
plt.show() 
```

```python
import sympy as sp
sp.init_printing()
from sympy import *

x = Symbol("x")
n = Rational("4")
a = Rational("0")
b = Rational("2")
hn= (b-a)/n
hn
```

```python
f=lambda x : Rational("3")*exp(-x)
x0=a; x1=x0+hn; x2=x1+hn; x3=x2+hn; x4=x3+hn
print(x0, x4)
f(x0), f(x1), f(x2), f(x3), f(x4)
```

```python
# (a) Riemann Summen
# Untersumme
rie_n = hn*f(x0)+hn*f(x1)+hn*f(x1)+hn*f(x2)
rie_n.evalf()
```

```python
# (b) Trapezregel
trapez_n = hn*(Rational("1/2")*f(x0)+f(x1)+f(x2)+f(x3)+Rational("1/2")*f(x4))
trapez_n.evalf()
```

```python
# (c) Simpson Verfahren
simp_n = hn/Rational("3")*(f(x0)+Rational("4")*(f(x1)+f(x3))+Rational("2")*(f(x2))+f(x4))
simp_n.evalf()
```
