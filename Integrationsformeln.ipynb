#### Funktionen ohne Stammfunktion $\frac{\sin(x)}{x}$, $\frac{\cos(x)}{x}$

```python
import math
import numpy as np              # array Funktionen
import matplotlib.pyplot as plt # plot Funktionen

x  = np.arange(-2*math.pi, 2*math.pi, 0.02) # x-Werte
yp = [ math.sin(x[i])/x[i] if x[i] > 0.0 else 0 for i in range(x.shape[0]) ] # Zugehörige y-Werte der Wertetabelle
ym = [ math.sin(x[i])/x[i] if x[i] < 0.0 else 0 for i in range(x.shape[0]) ] # Zugehörige y-Werte der Wertetabelle

plt.plot(x, np.array(yp)+np.array(ym), 'g-')       # Plot der Wertetabelle

plt.plot([0], [0], 'k.')
plt.plot([-2*math.pi, 2*math.pi], [0,  0], 'k-') 
plt.plot([0, 0], [-2,  2], 'k-') 
plt.axis([-2*math.pi, 2*math.pi, -2, 2])
```

#### Trapezregel

```python
import sympy as sp
sp.init_printing()
from sympy import *

# Variable
x = Symbol("x")
(x0, x1, h) = symbols("x0 x1 h")
(f0, f1) = symbols("f0 f1")
# Lagrange Interpolation mit 2 Punkten
l0 = (x-x1)/(x0-x1)
l1 = (x-x0)/(x1-x0)
lx = f0*l0+f1*l1
lx
```

```python
x1 = x0+h
l0 = (x-x1)/(x0-x1)
l1 = (x-x0)/(x1-x0)
lh = f0*l0+f1*l1
lh
```

```python
i = integrate(lh, (x, x0, x1))
i
```

```python
simplify(i)
```

#### Simpson-Verfahren

```python
import sympy as sp
sp.init_printing()
from sympy import *

# Variable
x = Symbol("x")
(x0,  x1,  x2, h) = symbols("x0 x1 x2 h")
(f0, f1, f2) = symbols("f0 f1 f2")
# Lagrange Interpolation mit 3 Punkten
l0 = ((x-x1)*(x-x2))/((x0-x1)*(x0-x2))
l1 = ((x-x0)*(x-x2))/((x1-x0)*(x1-x2))
l2 = ((x-x0)*(x-x1))/((x2-x0)*(x2-x0))
lx = f0*l0+f1*l1+f2*l2
lx
```

```python
x1 = x0+h
x2 = x1+h
l0 = ((x-x1)*(x-x2))/((x0-x1)*(x0-x2))
l1 = ((x-x0)*(x-x2))/((x1-x0)*(x1-x2))
l2 = ((x-x0)*(x-x1))/((x2-x0)*(x2-x0))
lh = f0*l0+f1*l1+f2*l2
#lh = lx.subs({"x1" : x0+h, "x2" : x1+h})
lh
```

```python
i = integrate(lh, (x, x0, x2))
i
```

```python
simplify(i)
```
